```python
import os
import sys

YOUR_REAL_API_KEY=""

os.environ["GOOGLE_API_KEY"] = YOUR_REAL_API_KEY
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma

try:
    from langchain_classic.chains import RetrievalQA
except ImportError:
    print("Error: Please run 'pip install langchain-classic'")
    sys.exit(1)

def buildChatbot(file_path):

    loader = PyPDFLoader(file_path)
    docs = loader.load()

    splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
    text = splitter.split_documents(docs)

    embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

    vectorstore = Chroma.from_documents(documents=text, embedding=embeddings)

    llm = ChatGoogleGenerativeAI(model="models/gemini-flash-latest", temperature=0.7)

    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=vectorstore.as_retriever()
    )

    return qa_chain


if __name__ == "__main__":

    if os.path.exists("data.pdf"):

        bot = buildChatbot("data.pdf")

        print("\n---  ChatBot is online ---")

        query = input("What do you have in your mind?: ")

        response = bot.invoke({"query": query})

        print("\nAnswer:", response["result"])

    else:
        print("Error: data.pdf not found")
```
